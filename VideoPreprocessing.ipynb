````
import cv2
def blend_img(x,y,z):
  image_data = [x,y,z]
  dst = image_data[0]
  for i in range(len(image_data)):
    if i == 0:
        pass
    else:
        alpha = 1.0/(i + 1)
        beta = 1.0 - alpha
        dst = cv2.addWeighted(image_data[i], alpha, dst, beta, 0.0)
  return dst
````

````
import cv2
import numpy as np
vidcap = cv2.VideoCapture('/content/sub5_alpha_nw.avi')
vidcap2=cv2.VideoCapture('/content/sub5_beta_nw.avi')
vidcap3=cv2.VideoCapture('/content/sub5_theta_nw.avi')
success,image = vidcap.read()
suc,img=vidcap2.read()
s,ig=vidcap3.read()
count = 0
img_array=[]
while success:
  image = cv2.resize(image,(200,200))
  img = cv2.resize(img,(200,200))
  ig=cv2.resize(ig,(200,200))
  ls=[image,img,ig]
  #avg_img = np.median(ls, axis=0)
  avg_img = np.mean(ls,axis=0)
  #avg_img = blend_img(image,img,ig)
  img_avg = avg_img.astype(np.uint8)
  #img_a = cv2.add(image,img)
  #img_avg = cv2.add(img_a,ig)
  whitee = np.zeros((39, 39, 3), np.uint8)
  whitee[:]=(255,255,255)
  img_avg[160:199,160:199]=whitee
  white1 = np.zeros((40, 20, 3), np.uint8)
  white1[:]=(255,255,255)
  img_avg[160:200,0:20]=white1
  img_array.append(img_avg)
  #cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  suc,img=vidcap2.read()
  s,ig=vidcap3.read()
  #print('Read a new frame: ', success)
  count += 1
````
